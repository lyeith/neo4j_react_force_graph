```python
import pandas as pd
```

```python
node_dct = {
    'address': 'data/paradise_papers.nodes.address.csv',
    'entity': 'data/paradise_papers.nodes.entity.csv',
    'intermediary': 'data/paradise_papers.nodes.intermediary.csv',
    'officer': 'data/paradise_papers.nodes.officer.csv',
    'other': 'data/paradise_papers.nodes.other.csv'
}

edges = 'data/paradise_papers.edges.csv'
```

```python
node_dtypes = {
    'country_codes': 'category',
    'countries': 'category',
    'sourceID': 'category',
    'valid_until': 'category',
    'note': 'category'
}

date_cols = ['inactivation_date']
```

```python
df_dct = {}

for k, v in node_dct.items():
   df_dct[k] = pd.read_csv(v, low_memory=False, dtype=node_dtypes)
```

# Node Data Structure Exploration

```python
list(df_dct['address'].name)
```

```python
from collections import Counter
import pprint

col_count = Counter()
col_dct = {}

for k, v in df_dct.items():
    col_dct[k] = list(v.columns)
    print(f'----{k}----')
    display(v.head(3))
    
    col_count += Counter(v.columns)
```

# Properties in each Dataset

```python
pprint.pp(col_dct)
```

# Common Properties across Datasets

```python
common_cols = {k:v for k, v in col_count.items() if v == len(df_dct)}
print(common_cols)
```
